#Pearson Correlation Threshold Tuning

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
import warnings

warnings.filterwarnings("ignore")

# Load Training Dataset
X_train = df.iloc[:, 1:1471]  # Features
Y_train = df.iloc[:, 0]       # Target Variable

# Compute Pearson Correlation Coefficients
correlation_matrix = X_train.corrwith(Y_train).abs()

# Settings
cv_folds = 10
best_overall_score = 0
best_model_name = None
best_threshold = None
best_selected_features = None

# Iterate over different threshold values
for quantile in np.arange(0.97, 1.00, 0.01):
    threshold = correlation_matrix.quantile(quantile)
    selected_features = correlation_matrix[correlation_matrix >= threshold].index
    X_selected = X_train[selected_features]

    print(f"\nEvaluating threshold = {quantile:.2f}")
    print(f" Selected {len(selected_features)} features: {list(selected_features)}\n")

    for name, model in [

    ('LR', LogisticRegression(max_iter=1000)),
    ('KNN', KNeighborsClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Gradient Boosting', GradientBoostingClassifier()),
    ('AdaBoost', AdaBoostClassifier()),
    ('Bagging', BaggingClassifier()),
    ('Extra Trees', ExtraTreesClassifier()),
    ('XGBoost',XGBClassifier(use_label_encoder=False, eval_metric='logloss'))
]:
        # Cross-validation
        cv_scores = cross_val_score(model, X_selected, Y_train, cv=cv_folds, scoring='accuracy')
        mean_accuracy = np.mean(cv_scores)
        print(f" {name}: CV Accuracy = {mean_accuracy:.4f}")

        # Update best results
        if mean_accuracy > best_overall_score:
            best_overall_score = mean_accuracy
            best_model_name = name
            best_threshold = quantile
            best_selected_features = selected_features

# Final Results
print("\n Best Model Summary")
print(f"Model: {best_model_name}")
print(f" Threshold: {best_threshold:.2f}")
print(f"Best CV Accuracy: {best_overall_score:.4f}")
print(f" Number of Selected Features: {len(best_selected_features)}")
print(f" Selected Features: {list(best_selected_features)}")
```

# Mutual Information Tuning

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
from sklearn.feature_selection import mutual_info_classif
import warnings

warnings.filterwarnings("ignore")

# Load Training Dataset
X_train = df.iloc[:, 1:1471]  # Features
Y_train = df.iloc[:, 0]       # Target Variable

# Compute Mutual Information Scores
mi_scores = mutual_info_classif(X_train, Y_train)
mi_sorted_indices = np.argsort(mi_scores)[::-1]  # Sort in descending order

# Define Models
models = [
    ('LR', LogisticRegression(max_iter=1000)),
    ('KNN', KNeighborsClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Gradient Boosting', GradientBoostingClassifier()),
    ('AdaBoost', AdaBoostClassifier()),
    ('Bagging', BaggingClassifier()),
    ('Extra Trees', ExtraTreesClassifier()),
    ('XGBoost', XGBClassifier(use_label_encoder=False, eval_metric='logloss'))
]

cv_folds = 10  # Use 10-fold cross-validation
best_overall_score = 0
best_num_features = None
best_model_name = None
best_selected_feature_names = None

# Iterate over number of selected features
for num_features in range(5, 31):
    print(f"\nEvaluating num_features={num_features}...\n")

    # Select Top 'num_features' based on MI Scores
    selected_features = mi_sorted_indices[:num_features]
    selected_feature_names = X_train.columns[selected_features]
    X_train_selected = X_train[selected_feature_names]

    print(f"Selected features ({num_features}): {list(selected_feature_names)}\n")

    for name, model in models:
        # Perform Cross-Validation
        cv_scores = cross_val_score(model, X_train_selected, Y_train, cv=cv_folds, scoring='accuracy')
        mean_accuracy = np.mean(cv_scores)

        print(f"num_features={num_features}, {name}: CV Accuracy = {mean_accuracy:.4f}")

        # Track best-performing model
        if mean_accuracy > best_overall_score:
            best_overall_score = mean_accuracy
            best_num_features = num_features
            best_model_name = name
            best_selected_feature_names = list(selected_feature_names)

# Final Results
print(f"\nBest Model: {best_model_name} with num_features={best_num_features}")
print(f"Best Cross-Validation Accuracy: {best_overall_score:.4f}")
print(f"Best Selected Features: {best_selected_feature_names}")
```

# Lasso Regression Tuning


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")

# Load Training Dataset
X_train = df.iloc[:, 1:1471]
Y_train = df.iloc[:, 0]
feature_names = df.columns[1:1471]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# Alpha values to try
alphas = np.logspace(-6, 0.5, 50)

# Results storage
results = []
best_score = 0
best_model = None
best_model_name = None
best_features = None

# Models
models = [
    ('LR', LogisticRegression(max_iter=1000)),
    ('KNN', KNeighborsClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Gradient Boosting', GradientBoostingClassifier()),
    ('AdaBoost', AdaBoostClassifier()),
    ('Bagging', BaggingClassifier()),
    ('Extra Trees', ExtraTreesClassifier()),
    ('XGBoost', XGBClassifier(use_label_encoder=False, eval_metric='logloss'))
]

for alpha in alphas:
    lasso = Lasso(alpha=alpha, max_iter=5000, random_state=42)
    lasso.fit(X_scaled, Y_train)

    selected_indices = np.where(lasso.coef_ != 0)[0]
    if 5 <= len(selected_indices) <= 30:
        selected_names = feature_names[selected_indices]
        X_selected = X_train[selected_names]

        print(f"\n Alpha: {alpha:.6f} | Selected Features: {len(selected_names)}")
        print(" Features:", list(selected_names))

        for name, model in models:
            scores = cross_val_score(model, X_selected, Y_train, cv=10, scoring='accuracy')
            mean_score = np.mean(scores)
            print(f" {name}: CV Accuracy = {mean_score:.4f}")

            if mean_score > best_score:
                best_score = mean_score
                best_model = model
                best_model_name = name
                best_features = selected_names

# Final Result
print("\n Best Model Summary:")
print(f" Model             : {best_model_name}")
print(f" Accuracy          : {best_score:.4f}")
print(f" Selected Features : {len(best_features)}")
print(f" Feature Names    : {list(best_features)}")

```
